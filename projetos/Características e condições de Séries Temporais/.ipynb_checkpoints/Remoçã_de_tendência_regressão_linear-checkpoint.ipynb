### Importando as bibliotecas

```python
# Técnica detrending
# Usando regressão linear para remover tendência

import pandas as pd
from pandas import Series 
import numpy as np
import matplotlib.pylab as plt
from datetime import datetime
%matplotlib inline
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6

# Biblioteca para criar o medelo de regreção linear
from sklearn.linear_model import LinearRegression

# Estacionariedade da série temporal
from statsmodels.tsa.stattools import adfuller
```

```python
dateparse = lambda dates: datetime.strptime(dates, '%Y-%m')
data = pd.read_csv('Dados//AirPassengers.csv', parse_dates=['Month'], index_col='Month', date_parser=dateparse)
```

```python
data.head()
```

### Criando modelo linear

```python
X = [i for i in range(0, len(data))]
X = np.reshape(X, (len(X), 1))
y = data.values
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
trend = model.predict(X)
```

```python
plt.plot(y)
plt.plot(trend)
plt.show()
```

Agora vamos subtrair da série temporal a tendência

```python
detrended = [y[i] - trend[i] for i in range(0, len(data))]
```

```python
plt.plot(detrended)
plt.show()
```

Logo a série tende a ser estacionária

```python
def adf_test(y):
    print('Resultado do teste Dickey-Fuller:')
    dftest = adfuller(y, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Teste', 'Valor P', 'Número de Lags', 'Número de observações'])
    for k, v in dftest[4].items():
        dfoutput['Valor crítico: ({})'.format(k)] = v
    print(dfoutput)
```

```python
y1 = data['#Passengers']
adf_test(y1)
```

```python
adf_test(detrended)
```
