### Importando as bibliotecas

```python
import pandas as pd
from pandas import Series
import numpy as np
import matplotlib.pylab as plt
from datetime import datetime
%matplotlib inline
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6

# Decomposição
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
dateparse = lambda dates: datetime.strptime(dates, '%Y-%m')
data = pd.read_csv('Dados//AirPassengers.csv', parse_dates=['Month'], index_col='Month', date_parser=dateparse)
```

```python
data.head()
```

### Visualização da série original

```python
plt.plot(data)
plt.title('Passageiros internacionais transportados', fontsize=24)
plt.xlabel('Data',fontsize=16)
plt.ylabel('Passageiros', fontsize=16)
plt.show()
```

### Fazendo a decomposição sazonal da série temporal

```python
result = seasonal_decompose(data, model='multiplicative')
```

```python
result.plot()
plt.show()
```

```python
result2 = seasonal_decompose(data, model='aditive')
```

```python
result2.plot()
plt.show()
```
