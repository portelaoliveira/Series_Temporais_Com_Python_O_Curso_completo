### Importando as bibliotecas 

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from datetime import datetime
%matplotlib inline
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6

# Funções de suavização
#from statsmodels.tsa.holtwinters import ExponentialSmoothing, SimpleExpSmoothing, Holt
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
```

```python
dateparse = lambda dates: datetime.strptime(dates, '%Y-%m')
data = pd.read_csv('Dados//AirPassengers.csv', parse_dates=['Month'], index_col='Month', date_parser=dateparse)
```

O Holt Winters é capaz de identificar os padrões de tendência e de sazonalida antes de fazer a previsão 

```python
# O período informado é o período em que os dados são obtidos, ou seja, anualment seasonal_periods=12

fit1 = ExponentialSmoothing(data, seasonal_periods=12, trend='additive', seasonal='additive').fit(use_boxcox=True)
```

```python
fit1.fittedvalues.plot(style='--', color='purple') # Modelo
```

```python
fit1.forecast(24).plot(style='--', marker='o', color='black', legend=True) # Escolhemos 2 anos forecast(24) para ser possível pegar o lemento sazonal e talvel a tendência
```

### Plotando o modelo junto com a previsão
