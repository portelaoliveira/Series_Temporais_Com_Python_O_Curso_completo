```python
# Importendo as bibliotecas

import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 13, 6 # Define as dimensões dos gráficos
```

```python
# Ações na bolsa de valores de 2006 à 2018

df = pd.read_csv('Dados//all-stocks-2006-01-01-to-2018-01-01.csv', parse_dates=['Date'])
```

```python
df.head()
```

```python
# Criando um intervalo de datas com o range

rng = pd.date_range(start='2006-01-01', end='2018-01-01', freq='B') # O'B' refere-se aos dias úteis
rng[~rng.isin(df.Date.unique())] # Pega os dias únicos, sem repetições
```

```python
# Selecionando a data do último mês de cada ano

sample_dates = pd.date_range(start='2006-01-01', end='2018-01-01', freq='B') # O'B' refere-se aos dias úteis
```

```python
year_end_dates = sample_dates[sample_dates.is_year_end]
year_end_dates
```

# Desempenhos pelo valor de **abertura** das ações na bolsa de valores

```python
# Os 5 piores desempenhos pelo valor de abertura das ações na bolsa de valores

worst_stockes = df[df.Date == df.Date.max()].sort_values('Open').head(5) # Ordem crescente
worst_stockes
```

```python
# Pegando os nomes 

ws = worst_stockes.Name.values
ws
```

```python
# Cópia de df

tdf = df.copy() # Deep copy
tdf = df.set_index('Date')
```

```python
# Filtrando

tdf[tdf.Name.isin(ws)].groupby('Name').Open.plot(title='Tendência histórica das 5 piores ações de 2017 na abertura', legend=True)
```

```python
# Os 5 melhores desempenhos pelo valor de abertura das ações na bolsa de valores

best_stockes = df[df.Date == df.Date.max()].sort_values('Open', ascending=False).head(5) # Ordem crescente
best_stockes
```

```python
# Pegando os nomes 

bs = best_stockes.Name.values
bs
```

```python
# Filtrando

tdf[tdf.Name.isin(bs)].groupby('Name').Open.plot(title='Tendência histórica das 5 melhores ações de 2017 na abertura', legend=True)
```
