```python
# Importendo as bibliotecas

import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6 # Define as dimensões dos gráficos
```

```python
# Terremotos na Itália no ano de 2016

df = pd.read_csv('Dados//italy_earthquakes_from_2016-08-24_to_2016-11-30.csv').set_index('Time') # DataFrame
df.index = pd.to_datetime(df.index) # Convertendo a colunda 'Time' para datetime
df.head()
```

##  Variação média da magnitude do terremoto a cada dia 

```python
'''
Estamos pegando o campo magnitude do terremoto, usamos resample('D') para regerar uma amostra usando o
campo data(D) e aplicamos nesse campo a média apply([np.mean]) e gerar o gráfico plot().
'''

df['Magnitude'].resample('D').apply([np.mean]).plot()
plt.title('Magnitude média por dia')
plt.xlabel('Tempo')
plt.ylabel('Magnitude')
```

## Variação média da magnitude do terremoto a cada dois dias 

```python
df['Magnitude'].resample('2D').apply([np.mean]).plot()
plt.title('Magnitude média a cada dois dias')
plt.xlabel('Tempo')
plt.ylabel('Magnitude')
```

## Filtrando os dados

```python
'''
Vamos filtrar os dados, para magnitude maior que três, vamos plotar a média móvel
'''
mag = df.loc[df['Magnitude'] >= 3.0, 'Magnitude']
```

```python
# Gerando o gráfico

pd.Series.rolling(mag, window=25).mean().plot(style='-r')
plt.title('Média móvel da magnitude')
plt.xlabel('Tempo')
plt.ylabel('Magnitude')
```
